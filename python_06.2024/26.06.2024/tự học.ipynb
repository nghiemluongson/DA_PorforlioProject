# Numpy

```python
import numpy as np

arr1 = np.array([1,2,0,8,9])

print(arr1)
```

```python
# kiểm tra kiểu dữ liệu của mảng
print(type(arr1))
```

```python
# kích thước của mảng
# cú pháp: <tên mảng>.shape

print(arr1.shape)
# arr1 đang là mảng 1 chiều => (5,) nghĩa là 5 phần tử
```

```python
# indexing: chỉ mục của mảng
# mảng có n phần tử => phần tử cuối cùng có index là n - 1

print(arr1[2])
# arr1: 1 2 0 8 9
# index:0 1 2 3 4
# => arr1[2]: in ra phần tử tại index = 2 của mảng => in ra 0
```

```python
# mảng 2 chiều

arr2 = np.array([[1,2,3],[4,5,6]])
print(arr2)
```

```python
print(arr2.shape)
# arr2 mảng 2 chiều, gồm 2 hàng 3 cột
```

```python
# truy xuất phần tử xong mảng n chiều
# cách 1: <tên mảng>[chỉ mục hàng][chỉ mục cột]
# cách 2: <tên mảng>[chỉ mục hàng, chỉ mục cột]

arr2[1][2]
# in ra phần tử tại hàng thứ 2, cột thứ 3
```

```python
arr2[1,2]
```

```python
# chỉ mục lớn nhất hàng / cột có thể xem như -1
# muốn truy xuất trước đó (lùi về) thì cứ -2 -3 ...

arr2[1][-1]
# in phần tử hàng thứ 2, ở cột cuối cùng (-1)
```

```python
# np.arange(value1, value2, step): tạo ra 1 mảng tăng dần từ value1 -> value2
# nếu ko có step => auto cách nhau 1 đơn vị
# step = 2: cách nhau 2 đơn vị
# value1 <= giá trị < value2: ko nhận giá trị value2 trong mảng

arr4 = np.arange(0,12,2)
arr4
```

```python
# np.linspace(value1, value2, num): tạo ra 1 mảng tăng dần từ value1 -> value2 với num giá trị
# khoảng cách giữa 2 phần tử kế nhau = value2 / num

arr5 = np.linspace(0,12,20)
arr5
```

```python
# intersect: liệt kê ra các phần tử giống nhau có trong 2 mảng
# union: các giá trị khác nhau
# 

arr21 = np.array([1,2,0,8,9])
arr22 = np.array([1,2,0,8,9,12,4,4,5,2,4,5])
print(np.intersect1d(arr21,arr22))
print(np.union1d(arr21,arr22))
print(np.setdiff1d(arr21,arr22))
print(np.setxor1d(arr21,arr22))
```
