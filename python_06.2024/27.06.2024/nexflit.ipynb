```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import chardet

%matplotlib inline
sns.set(color_codes=True)
```

```python
# Try reading the file with different encodings
file_path = 'netflix_titles.csv'

# Example with 'latin1' encoding
try:
    df = pd.read_csv(file_path, encoding='latin1')
except UnicodeDecodeError:
    print("Encoding 'latin1' failed. Trying a different encoding...")

# Example with 'iso-8859-1' encoding
try:
    df = pd.read_csv(file_path, encoding='iso-8859-1')
except UnicodeDecodeError:
    print("Encoding 'iso-8859-1' failed. Trying a different encoding...")

# Example with 'cp1252' encoding
try:
    df = pd.read_csv(file_path, encoding='cp1252')
except UnicodeDecodeError:
    print("Encoding 'cp1252' failed. Trying a different encoding...")

# Detect the encoding
with open(file_path, 'rb') as f:
    result = chardet.detect(f.read())
    encoding = result['encoding']

# Read the file with the detected encoding
df = pd.read_csv(file_path, encoding=encoding)
df
```
